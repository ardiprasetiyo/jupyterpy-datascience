<h2>Quiz 2 : Built-in Function</h2>

Jawab Pertanyaan di bawah ini :
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

Answer :  <br> <br>
<b> Built-in Function </b> adalah sekumpulan fungsi yang sudah disediakan oleh bahasa Python itu sendiri <br>
Beberapa <b>contoh built-in function</b> diantaranya max(), min(), sum()

```python
# Contoh penggunaan built in function
ex_list = [2,4,5]
max(ex_list) # fungsi max untuk mengembalikan item dengan nilai terbesar pada konteks list
min(ex_list) # fungsi min untuk mengembalikan item dengan nilai terkecil pada konteks list
sum(ex_list) # fungsi sum untuk mengembalikan hasil penjumlahan seluruh item pada list
```

<h2>Quiz 6 : Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]
def mean_list(inp_list):
    return sum(inp_list)/len(inp_list)
    
print(mean_list(obj_list))
```

Expected Output :

13.9

<h2>Quiz 7 : Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_pengali = [1, 2, 3]
def kali_list(obj_list, obj_pengali):
    return obj_list + obj_pengali
    
    
print(kali_list(obj_list, obj_pengali))
```

Expected Output :

[2, 4, 5, 6, 1, 2, 3]
